```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
```

## Dias

```python
# importar dataset
df=pd.read_csv("dailyActivity_merged.csv")
# vista previa dataset
df.head(5)
```

```python
# Eliminar columnas
df=df.drop(["ModeratelyActiveDistance","LightActiveDistance","SedentaryActiveDistance","VeryActiveMinutes","TrackerDistance","LoggedActivitiesDistance","VeryActiveDistance", "FairlyActiveMinutes","LightlyActiveMinutes","SedentaryMinutes"], axis=1)
# Ver informacion de dataset
df.info()
```

```python
# loop para obtener el dia de la semana a partir de la fecha
for lab, row in df.iterrows():

    fecha_obj = datetime.strptime(row["ActivityDate"], "%m/%d/%Y")
    df.loc[lab, "Day"]=fecha_obj.strftime("%A")
```

```python
# Obtener la media y sumatoria de la disctancia recorrida por dia
df_dist =df.groupby("Day")["TotalDistance"].agg([np.sum])
df_dist=df_dist.sort_values(by="sum", ascending=False)
```

```python
ax=df_dist.plot(kind="bar", rot=45, figsize=(10, 5),color=["blue"])
plt.title("Distancia recorrida por dia de la semana")
plt.xlabel("Dias de la semana")
plt.ylabel("Distancia")
plt.legend(["Distancia total"])
ax.grid(axis = "y")
ax.locator_params(axis='y', nbins=10)
ax.set_ylim(0, 1000)
plt.show()
```

```python
# Obtener la media y sumatoria de los pasos totales por dia
df_steps=df.groupby("Day")["TotalSteps"].agg([np.sum])
df_steps=df_steps.sort_values(by="sum", ascending=False)
```

```python
ax=df_steps.plot(kind="bar", color=["green"], rot=45, figsize=(10, 5))
plt.title("Numero de pasos por dia de la semana")
plt.xlabel("Dias de la semana")
plt.ylabel("Numero de pasos")
plt.legend(["Suma de Pasos"])
ax.grid(axis = "y")
ax.locator_params(axis='y', nbins=10)
ax.set_ylim(0, 1400000)
plt.show()
```

## Horas

```python
# Cargar Dataset
df_hour=pd.read_csv("hourlySteps_merged.csv")
# preview dataset
df_hour.head(5)
```

```python
# obtener infomacion sobre el tipo de variables en el dataset
df_hour.info()
```

```python
# agregar una nueva columna de horas solo con la hora
df_hour["Hour"]=df_hour["ActivityHour"].apply(lambda x:x[-10:])
```

```python
# 3limina columna "ActivityHour"
df_hour=df_hour.drop("ActivityHour", axis=1)
```

```python
# Agrupar por hora y sumar numero de pasos
df_hour=df_hour.groupby("Hour")["StepTotal"].sum()
# Ordenar las horas populares de mayor a menor
df_hour=df_hour.sort_values(ascending=False)
```

```python
ax=df_hour.plot(kind="barh", color=["red"],figsize=(10, 5))
plt.title("Numero de pasos por hora")
plt.xlabel("Horas del dia")
plt.ylabel("Numero de pasos")
plt.legend(["Suma de Pasos"])
ax.grid(axis = "x")
ax.locator_params(axis='x', nbins=15)
plt.show()
```

## Calorias por dia

```python
# importar dataset
df=pd.read_csv("dailyCalories_merged.csv")
# vista previa dataset
df.head(5)
```

```python
# loop para obtener el dia de la semana a partir de la fecha
for lab, row in df.iterrows():

    fecha_obj = datetime.strptime(row["ActivityDay"], "%m/%d/%Y")
    df.loc[lab, "Day"]=fecha_obj.strftime("%A")
```

```python
# Obtener la media y sumatoria de la disctancia recorrida por dia
df_dist =df.groupby("Day")["Calories"].agg([np.sum])
df_dist=df_dist.sort_values(by="sum", ascending=False)
```

```python
#Type plot, rotation of bin labels, size plot, color plot
ax=df_dist.plot(kind="bar", rot=45, figsize=(10, 5), color=["orange"])
# title plot
plt.title("Calorias totales por dia de la semana")
# x label
plt.xlabel("Dias de la semana")
# y label
plt.ylabel("Calorias")
# description legend
plt.legend(["Calorias totales"])
# put grins in axis y
ax.grid(axis = "y")
# number of bins in y
ax.locator_params(axis='y', nbins=10)
# change scale in y
ax.set_ylim(0, 400000)
plt.show()
```

## Calorias por hora

```python
# importar dataset
df=pd.read_csv("hourlyCalories_merged.csv")
# vista previa dataset
df.head(5)
```

```python
# agregar una nueva columna de horas solo con la hora
df["Hour"]=df["ActivityHour"].apply(lambda x:x[-10:])
```

```python
# elimina columna "ActivityHour"
df=df.drop("ActivityHour", axis=1)
```

```python
# Agrupar por hora y sumar numero de pasos
df=df.groupby("Hour")["Calories"].sum()
# Ordenar las horas populares de mayor a menor
df=df.sort_values(ascending=False)
```

```python
ax=df.plot(kind="barh", color=["purple"],figsize=(10, 5))
plt.title("Numero de calorias por hora")
plt.xlabel("Horas del dia")
plt.ylabel("Numero de Calorias")
plt.legend(["Suma de Calorias"])
ax.grid(axis = "x")
ax.locator_params(axis='x', nbins=10)
plt.show()
```

```python
# Los dias donde mas distancias se recorren son martes, miercoles, jueves, sabado
# los dias con mas de 1 millon de pasos son martes, miercoles, jueves y sabado
# los dias donde mas se consumen mas de 300 mil calorias son martes, miercoles y jueves

# las horas en las cuales se tienen mas de 400 mil pasos son 2,1,6,7,5,4 pm y 1 y 12 am
# las horas donde mas de 100 mil colorias se consumieron son 2,1,6,5,7,4 pm y 12,2,1 am
```

1.	¿Cuáles son algunas tendencias de uso de los dispositivos inteligentes?
* Los dias donde mas distancias se recorrio, mas pasos se dieron y mas calorias se quemaron fueron martes, miercoles y jueves
* Las horas donde mas pasos se dieron y mas calorias se consumieron fueron 1,2,4,5,6,7 pm y 12, 1 am

2.	¿Cómo se podrían aplicar estas tendencias a los clientes de Bellabeat?
* Estos patrones podrian servir para investigar mas a fondo el perfil de las  personas, quiza sean personas que trabajan en homeoffice, o que tienen gimnacios cerca de su empleo, o quiza los usuarios eran amas de casa, estos resultados nos dicen que se realiza mas actividades fisicas entre semana y por las tardes

3.	¿Cómo podrían ayudar estas tendencias a influir en la estrategia de marketing de Bellabeat?
* Viendo que entre semana y por las tardes es cuando mas actividades realizan, entonces se podrian lanzar publicidad  de productos justo en las tardes y entre semana
* Se Podrian usar esos horarios y dias para lanzar promociones
* Se podrian ofrecer productos exclusivos para su uso en las tardes, como bloqueador solar, gorras, lentes de sol, etc
